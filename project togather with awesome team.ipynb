```python
page = requests.get("https://www.imdb.com/list/ls066055776/")
print(page)
soup = BeautifulSoup(page.content, "html.parser")
drama= soup.find_all('h3', class_= "lister-item-header")
print (drama)
```

```python
movie_title=[]
for m in drama:
    movie_title.append(m.text)
print(movie_title)
```

```python
page = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
#print(page)

soup=BeautifulSoup(page.content, "html.parser")
drama=soup.find_all('div', class_ ="inline-block ratings-imdb-rating")
#print(drama)

movie_ratings=[]
for m in drama:
    movie_ratings.append(m.text)
#print(movie_ratings)


m_ratings1=[]
for i in range (len(movie_ratings)):
    new_movie_ratings=re.sub("[A-Za-z]",' ',movie_ratings[i])
    m_ratings1.append(new_movie_ratings)
    m_ratings1[i]=m_ratings1[i].strip()

#print(m_ratings1)
df = pd.DataFrame ({'Name of Movie' : m_titles, 'Year launched': years, 'Ratings': m_ratings1})
df
```

```python
page = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
#print(page)

soup=BeautifulSoup(page.content, "html.parser")
drama=soup.find_all('span', class_="runtime")
#print(drama)

movie_duration=[]
for m in drama:
    movie_duration.append(m.text)
#print(movie_duration)


m_duration=[]
for i in range (len(movie_duration)):
    new_movie_duration=re.sub("[^0-9]",' ',movie_duration[i])
    m_duration.append(new_movie_duration)
    m_duration[i]=int(m_duration[i].strip())

#print(m_duration)

#EXTRAS : Flavor Text for giving time units
duration_unit = ['min'] * 50
data = {'duration':m_duration, 'unit':duration_unit}

#print(len(duration_unit))
#print(len(m_duration))




df_duration = pd.DataFrame(data)

print(df_duration)
```

```python
page = requests.get("https://www.imdb.com/list/ls066055776/?sort=list_order,asc&st_dt=&mode=detail&page=1&ref_=ttls_vm_dtl")
print(page)
soup = BeautifulSoup(page.content, "html.parser")
drama= soup.find_all('p', class_
                    ="text-muted text-small")
#print (drama)
actors = []

for a in drama:
    actors.append(a.text)
#print(actors)
stars = []

for i in range (len(actors)):
    new_actors = re.sub("[^A-Za-z]", ' ', actors[i])
    stars.append(new_actors)
    stars[i]=stars[i].strip()

stars = stars[:50]
df = pd.DataFrame (stars, columns = ['Name of Actors'])
df
```
