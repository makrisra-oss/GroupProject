```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
page = requests.get("https://www.imdb.com/list/ls066055776/")
print(page)
soup = BeautifulSoup(page.content, "html.parser")
drama= soup.find_all('h3', class_= "lister-item-header")
print (drama)
```

```python
movie_title=[]
for m in drama:
    movie_title.append(m.text)
print(movie_title)
```

```python
# we are finding the movie details
year_launched= soup.find_all('span', class_="lister-item-year text-muted unbold")
#year_launched


years = []
for year in year_launched:
    years.append(year.text)
    
years = years[:50]
    
    
#     year1 = soup.find('span', class_="lister-item-year text-muted unbold").text 
#     years.append (year1)
# #print (years)
df = pd.DataFrame ({'Name of Movie' : m_titles, 'Year launched': years})
df


```

```python
page = requests.get("https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc")
#print(page)

soup=BeautifulSoup(page.content, "html.parser")
drama=soup.find_all('div', class_ ="inline-block ratings-imdb-rating")
#print(drama)

movie_ratings=[]
for m in drama:
    movie_ratings.append(m.text)
#print(movie_ratings)


m_ratings1=[]
for i in range (len(movie_ratings)):
    new_movie_ratings=re.sub("[A-Za-z]",' ',movie_ratings[i])
    m_ratings1.append(new_movie_ratings)
    m_ratings1[i]=m_ratings1[i].strip()

#print(m_ratings1)
df = pd.DataFrame ({'Name of Movie' : m_titles, 'Year launched': years, 'Ratings': m_ratings1})
df
```
